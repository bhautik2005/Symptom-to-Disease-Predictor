# # Symptom Co-occurrence Patterns

```python
symptom_cols = [c for c in df.columns if c != 'diseases']
disease_counts = df['diseases'].value_counts()
```

```python
# ── 5A. Symptom co-occurrence matrix ──────────────────────────────────────────
# For every pair of symptoms, count how many records have BOTH = 1
# Use dot product: (binary_matrix).T @ (binary_matrix)

symptom_matrix = df[symptom_cols].astype(int)
co_matrix = symptom_matrix.T.dot(symptom_matrix)  # shape: (symptoms x symptoms)

# Zero out the diagonal (self-co-occurrence is not useful)
np.fill_diagonal(co_matrix.values, 0)

# Keep only top 30 symptoms for readable heatmap
top30 = df[symptom_cols].sum().sort_values(ascending=False).head(30).index
co_top = co_matrix.loc[top30, top30]

plt.figure(figsize=(16, 13))
sns.heatmap(
    co_top,
    cmap="Blues",
    linewidths=0.2,
    xticklabels=True,
    yticklabels=True,
    annot=False
)
plt.title("Symptom Co-occurrence Matrix (top 30 symptoms)", fontsize=13)
plt.xticks(rotation=45, ha='right', fontsize=7)
plt.yticks(fontsize=7)
plt.tight_layout()
plt.savefig("symptom_cooccurrence.png", dpi=150)
plt.show()

# Print top 10 most co-occurring pairs
print("--- Top 10 Most Co-occurring Symptom Pairs ---")
# Get upper triangle to avoid duplicate pairs
upper = co_matrix.where(np.triu(np.ones(co_matrix.shape), k=1).astype(bool))
top_pairs = (
    upper.stack()
    .reset_index()
    .rename(columns={'level_0': 'Symptom_A', 'level_1': 'Symptom_B', 0: 'co_count'})
    .sort_values('co_count', ascending=False)
    .head(10)
)
print(top_pairs.to_string(index=False))
```

```python
# ── 5B. Signature symptom sets per disease ────────────────────────────────────
# For each of the top 10 diseases, find the 5 most "signature" symptoms
# Signature = high average presence FOR that disease AND low average elsewhere

top10_diseases = disease_counts.head(10).index.tolist()

print("\n--- Signature Symptoms per Disease ---")
for disease in top10_diseases:
    in_disease  = df[df['diseases'] == disease][symptom_cols].mean()
    out_disease = df[df['diseases'] != disease][symptom_cols].mean()

    # Score: how much MORE common in this disease vs others
    score = in_disease - out_disease
    top5 = score.sort_values(ascending=False).head(5)

    print(f"\n{disease}")
    for sym, val in top5.items():
        print(f"   {sym:<45} +{val:.2f}")
```

```python
# ── 5C. Visual: Top 5 signature symptoms for top 6 diseases ───────────────────
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, disease in enumerate(top10_diseases[:6]):
    in_d  = df[df['diseases'] == disease][symptom_cols].mean()
    out_d = df[df['diseases'] != disease][symptom_cols].mean()
    score = (in_d - out_d).sort_values(ascending=False).head(8)

    axes[i].barh(score.index[::-1], score.values[::-1], color='tomato')
    axes[i].set_title(disease, fontsize=10, fontweight='bold')
    axes[i].set_xlabel("Signature score", fontsize=8)
    axes[i].tick_params(axis='y', labelsize=7)
    axes[i].axvline(0, color='gray', linewidth=0.8)

plt.suptitle("Signature Symptoms per Disease (top 6 diseases)", fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig("signature_symptoms.png", dpi=150, bbox_inches='tight')
plt.show()
```

# @ Feature Engineering & Train/Test Split

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import joblib
import os
```

```python
# Create the 'outputs' directory if it doesn't exist
os.makedirs('outputs', exist_ok=True)

# ── 6A. Confirm data types ─────────────────────────────────────────────────────
symptom_cols = [c for c in df.columns if c != 'diseases']
print("Total symptom features:", len(symptom_cols))

# Your dataset already has binary columns — just make sure they are int 0/1
df[symptom_cols] = df[symptom_cols].astype(int)

# Check: any non-0/1 values?
unique_vals = df[symptom_cols].stack().unique()
print("Unique values in symptom columns:", unique_vals)  # should be [0, 1]
```

```python

# ── 6B. Handle any remaining issues ───────────────────────────────────────────
# Drop duplicates
before = len(df)
df.drop_duplicates(inplace=True)
print(f"\nDropped {before - len(df)} duplicate rows. Remaining: {len(df)}")

# Drop rows where disease label is missing
df.dropna(subset=['diseases'], inplace=True)

# Strip whitespace from disease names
df['diseases'] = df['diseases'].str.strip()

# --- FIX: Remove classes with only 1 sample for stratified split ---
disease_counts = df['diseases'].value_counts()
single_sample_diseases = disease_counts[disease_counts == 1].index.tolist()
if single_sample_diseases:
    df = df[~df['diseases'].isin(single_sample_diseases)]
    print(f"Removed {len(single_sample_diseases)} disease classes with only 1 sample for stratified split.")

# IMPORTANT: Create a fresh copy of the DataFrame to avoid SettingWithCopyWarning
df = df.copy()
```

```python

# ── 6C. Encode the target label ────────────────────────────────────────────────
le = LabelEncoder()
df['disease_encoded'] = le.fit_transform(df['diseases'])

print(f"\nTotal disease classes: {len(le.classes_)}")
print("First 10 class labels:", list(le.classes_[:10]))

# Save the label encoder — you need it later to decode predictions
joblib.dump(le, "outputs/label_encoder.pkl")
print("✅ Label encoder saved → outputs/label_encoder.pkl")
```

```python


# ── 6D. Optional: Feature selection (remove low-variance symptoms) ─────────────
# Symptoms that appear in < 0.5% of records add noise, not signal
threshold = 0.005
symptom_freq = df[symptom_cols].mean()
useful_symptoms = symptom_freq[symptom_freq >= threshold].index.tolist()
removed = len(symptom_cols) - len(useful_symptoms)
print(f"\nFeature selection: removed {removed} near-zero symptoms (< {threshold*100}% frequency)")
print(f"Symptoms remaining: {len(useful_symptoms)}")

# Save the useful symptom list — needed for inference later
pd.Series(useful_symptoms).to_csv("outputs/feature_columns.csv", index=False)
print("✅ Feature columns saved → outputs/feature_columns.csv")
```

```python
# ── 6E. Define X and y ────────────────────────────────────────────────────────
X = df[useful_symptoms].values
y = df['disease_encoded'].values

print(f"\nX shape: {X.shape}   (rows x features)")
print(f"y shape: {y.shape}   (rows)")
```

```python
# ── 6F. Train / test split — stratified ───────────────────────────────────────
# stratify=y ensures every disease class appears in both train and test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(f"\nTrain size : {X_train.shape[0]} rows")
print(f"Test size  : {X_test.shape[0]} rows")
```

```python

# ── 6G. Save splits ───────────────────────────────────────────────────────────
np.save("outputs/X_train.npy", X_train)
np.save("outputs/X_test.npy",  X_test)
np.save("outputs/y_train.npy", y_train)
np.save("outputs/y_test.npy",  y_test)

# Also save a clean CSV version of the full processed dataset
df_clean = df[['diseases', 'disease_encoded'] + useful_symptoms]
df_clean.to_csv("outputs/cleaned_dataset.csv", index=False)

print("\n✅ Phase 6 complete. Saved:")
print("   outputs/X_train.npy")
print("   outputs/X_test.npy")
print("   outputs/y_train.npy")
print("   outputs/y_test.npy")
print("   outputs/cleaned_dataset.csv")
print("   outputs/label_encoder.pkl")
print("   outputs/feature_columns.csv")
```

```python
# ── 6H. Quick sanity check ────────────────────────────────────────────────────
print("\n--- Sanity Check ---")
print("Train class distribution (first 5):")
unique, counts = np.unique(y_train, return_counts=True)
for cls, cnt in list(zip(le.inverse_transform(unique), counts))[:5]:
    print(f"   {cls:<40} {cnt}")
```
